# Toy datasets
scikit-learnには標準で小規模データセットがいくつか入っています。


### Boston house-prices (ボストン市の住宅価格)
米国ボストン市郊外における地域別の住宅価格のデータセット。

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(dir(boston))
print(boston.feature_names)
print(boston.data.shape)
```

### Iris (アヤメの計測データ、通称：アイリス)
“setosa”, “versicolor”, “virginica” という 3 種類の品種のアヤメのがく片 (Sepal)、花弁 (Petal) の幅および長さを計測したデータです。

```python
from sklearn.datasets import load_iris
data = load_iris()
data.target[[10, 25, 50]]
list(data.target_names)
```

### Diabetes (糖尿病患者の診断データ)
糖尿病患者 442 人の検査数値と 1 年後の疾患進行状況（正規化済み）。

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
X = diabetes.data[:50]
y = diabetes.target[:50]
print(X.shape)
print(y.shape)
print(dir(diabetes))
print(diabetes.data.max())
```

### Digits (数字の手書き文字)
0 ～ 9 の 10 文字の手書きの数字を 64 (8×8) 個の画素に分解したものです。

```python
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)
print(type(digits.images[0]))

import matplotlib.pyplot as plt
plt.gray()
plt.matshow(digits.images[0]) 
plt.show() 
```

### Linnerud (生理学的特徴と運動能力の関係)
ノースカロライナ州立大学の A. C. linnerud 博士が作成した、20 人の成人男性に対してフィットネスクラブで測定した 3 つの生理学的特徴と 3 つの運動能力の関係。

```python
from sklearn.datasets import load_linnerud
linnerud = load_linnerud()
print(dir(linnerud))

import pandas as pd
pd.DataFrame(linnerud.data, columns=linnerud.feature_names).head()
#pd.DataFrame(linnerud.target, columns=linnerud.target_names).head()
```
